```python
#training
#i.load the DATA(MLP)
import pandas as pd
file=pd.read_csv("/content/13 MLP winequality-red.csv")
x=file[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol" ]]
y=file[["quality"]]
#cross validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)

#2.load the model(MLP)
from sklearn.neural_network import MLPClassifier


hidden_layers=[
    (5,2,),
    (6,2,),
    (7,),
    (8,2,),
    (9,1,7,),
]
models=[]
#fit data into model

for layer in hidden_layers:
  model=MLPClassifier(hidden_layer_sizes=layer)
  model.fit(xtrain,ytrain)
  models.append(model)

#validation
from sklearn.metrics import accuracy_score
for i in range(len(models)):
  op=models[i].predict(xtest)
  score=accuracy_score(ytest,op)
  print("Accuracy=",score*100)


#testing
test_data=[[00.5,0.56,00.33,00.555,0.99,0.99,0.1,0.75,0.22,0.98,0.56]]
for i in range(len(models)):
  res=models[i].predict(test_data)
  print(res)
```
